```python
import numpy as np
from scipy.stats import skewnorm, bootstrap, monte_carlo_test

n=100
x=skewnorm.rvs(a=1,size=n,random_state=42)

def kurt(x):
    mu=np.mean(x)
    sigma=np.sqrt(np.var(x))
    return np.mean(((x-mu)/sigma)**4)

def gausian_rv(x):
    B=1000
    kurt_g=[]
    for i in range(B):
        xg=np.random.normal(size=100)     ## for gausian you have to use normal
        kurt_g.append(kurt(xg))
    
    kurt_g= np.array(kurt_g)
    lower=np.percentile(kurt_g,2.5)
    upper=np.percentile(kurt_g,97.5)

    if lower<=kurt(x)<=upper:                    ##
        print("accept h0")
    else:
        print("reject H0")

    manual_p=np.mean(np.abs(kurt_g-3)>=abs(kurt(x)-3))    ##
    print("manual_p",manual_p)

    #Monte Carlo
    res_mc = monte_carlo_test(
    x,
    rvs=lambda size: np.random.normal(size=size),
    statistic=kurt,                                ##
    alternative='two-sided',
    n_resamples=B)

    print("p_mc",res_mc.pvalue)

gausian_rv(x)

```

```python
def boot(x):
    B=1000
    kurt_b=[]
    for i in range(B):
        xb=np.random.choice(x,size=100,replace=True)
        kurt_b.append(kurt(xb))
    kurt_b=np.array(kurt(xb))

    lower=np.percentile(kurt_b,2.5)
    upper=np.percentile(kurt_b,97.5)

    if lower<=kurt(x)<=upper:
        print("accept h0")
    else:
        print("reject H0")

    pval_manual = np.mean(np.abs(kurt_b - np.mean(kurt_b)) >= abs(kurt(x) - np.mean(kurt_b)))        ##
    print("Manual p-value =", pval_manual)

    #Monte Carlo
    res_mc = monte_carlo_test(
    x,
    rvs=lambda size: np.random.normal(size=size),
    statistic=kurt,                                ##
    alternative='two-sided',
    n_resamples=B)

    print("p_mc",res_mc.pvalue)
    

boot(x)
```
