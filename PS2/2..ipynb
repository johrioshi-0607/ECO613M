```python
import math
from scipy.stats import fisher_exact

a=8
b=2
c=3
d=9
n=a+b+c+d

def prob_table(a,b,c,d,n):
    r1 = a+b
    r2 = c+d
    c1 = a+c
    return math.comb(r1,a)*math.comb(r2,c1-a)/math.comb(n,c1)

def p_value(a,b,c,d,n):

    r1 = a+b
    r2 = c+d
    c1 = a+c

    PmyT = prob_table(a,b,c,d,n)

    amin = max(0,c1-r2)
    amax = min(r1,c1)

    pvalue = 0

    for i in range(amin,amax+1):

        na = i
        nb = r1 - na
        nc = c1 - na
        nd = r2 - nc

        table = prob_table(na,nb,nc,nd,n)

        if table <= PmyT + 1e-12:
            pvalue += table

    return pvalue

pval = p_value(a,b,c,d,n)
print("Manual =", pval)

p = fisher_exact([[a,b],[c,d]], alternative='two-sided')[1]
print("Scipy =", p)
```
