```python
import numpy as np
q=3
coeff=np.array([0.5,0.5,-0.2])

x=np.zeros(10000)

x[0]=0
x[1]=0
x[2]=0

for t in range(3,10000): 
    x[t]=coeff[0]*x[t-1]+coeff[1]*x[t-2]+coeff[2]*x[t-3]+np.random.normal(0.5,2**(1/2))

print(x[0:10])
```

```python
meanT = 0.5 / (1 - np.sum(coeff))
print("mean =", np.mean(x))
print("meanT =", meanT)

def autocov(x, h):
    m = np.mean(x)
    
    if h == 0:
        return np.mean((x - m)**2)   
    
    return np.mean((x[h:] - m) * (x[:-h] - m))

p = []
for l in range(6):
    p.append(autocov(x, l))

print("autocovar =", p)
print("autocovarT =", autocov)
```
