```python
import numpy as np
import pandas as pd

df = pd.read_csv("bjm.csv")
x = df.iloc[:,0].values

mu=np.mean(x)

def autocov(x,h):
    m=np.mean(x)
    if h==0:
        return np.mean((x-m)**2)
    else:
        return np.mean((x[h:]-m)*(x[:-h]-m))

AIC=[]
coeff=[]
sigma_sq=[]
for p in range(1,11):
    #autocovariance
    auto=[]
    for i in range(p+1):
        auto.append(autocov(x,i))

    #yule walker matrix
    G=np.zeros((p,p))
    for j in range(p):
        for k in range(p):
            G[j,k]=auto[abs(j-k)]
    #solve for coeff
    rhs=auto[1::]
    a=np.linalg.solve(G,rhs)

    coeff.append(a)
    
    #sigma square (noise variance)
    sigma2=auto[0]-np.dot(rhs,a)
    sigma_sq.append(sigma2)
    
    #AIC
    aic  = len(x) * np.log(sigma2) + 2 * p

    AIC.append(aic)

#print p
best_p = np.argmin(AIC) + 1
print("p =", best_p)

#print coeff
best_a=coeff[np.argmin(AIC)]
print("coeff=",best_a)


noise_mean=mu*(1-sum(best_a))
print("mean=",noise_mean)

best_sigma2=sigma_sq[np.argmin(AIC)]
print("var=",best_sigma2)
```
