```python
import numpy as np
q=6
coeff=np.array([6,4,7,3,2,3])

import math

err=np.random.normal(0.5,math.sqrt(2),size=10000+q)

X=[]

for t in range(10000):
    Xt=err[t+q]
    for i in range(1,q+1):
        Xt+=coeff[i-1] * err[t+q-i]
    X.append(Xt)

X[0:5]
```

```python
meanT=0.5/(1-coeff[0]-coeff[1]-coeff[2])
print("mean=",np.mean(X))
print("meanT=",meanT)

varT=2*(1+sum(coeff**2))
print("var=",np.var(X))
print("varT=",varT)

auto=[]
a=np.insert(coeff,0,1)
for j in range(1,q+3):
    ac=0
    if j<=q:
        for k in range(len(a)-j):
            ac+=a[k]*a[k+j]
        auto.append(2*ac)
    else:
        auto.append(0)

def autocov(X, h):
    m = np.mean(X)
    return np.mean((X[h:] - m) * (X[:-h] - m))

p=[]
for l in range(1,q+3):
    p.append(autocov(X,l))

print("autocovar=",p)
print("autocovarT=",auto)
```
