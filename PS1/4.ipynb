```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array(np.random.rand(1000))

beta=int(input("enter beta: "))

y=[]
for i in x:
    c=0
    for j in range(3):
        a=np.random.exponential(beta)
        c+=a
    y.append(c)
    

#KS Test stat
y.sort()

Fn=[]
Fx=[]

for i in range(1000):
    cdf=(i+1)/1000
    Fn.append(cdf)
    x = y[i]
    cdfx = 1 - np.exp(-x/beta) * (1 + x/beta + (x/beta)**2 / 2)
    Fx.append(cdfx)


D=[]
for i in range(1000):
    D.append(np.abs(Fx[i]-Fn[i]))

Dn=np.max(D)

KS_test=(1000**(1/2))*Dn

print("KS_test",KS_test)

pvalue = 0
for k in range(1, 51):
    pvalue += 2*((-1)**(k-1))*np.exp(-2*(k**2)*(KS_test**2))

print("p-value =", pvalue)


if pvalue>0.05:
    print("Data follows Gamma(3,beta)")
else:
    print("Reject fit")

```
