```python
#poisson(lambda)

import numpy as np
import matplotlib.pyplot as plt
lambdaa=float(input("enter lambda: "))
y=[]
for i in range(1000):
    x=[]
    count=0
    a=np.random.exponential(1/lambdaa)          #imp
    x.append(a)
    while sum(x)<=1:
        count+=1
        a=np.random.exponential(1/lambdaa)
        x.append(a)
    y.append(count)

print(y[:10])

plt.hist(y, bins=5, density=True)
plt.show()
        
        
```
