```python
#binomial(n,p)
import numpy as np
import matplotlib.pyplot as plt
import math
n=int(input("enter n: "))
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    x=[]
    for j in range(n):
        a=np.random.rand()
        if a<p:
            x.append(1)
        else:
            x.append(0)
    y.append(x.count(1))

#Test stat
y1=np.array(y)
xi=np.unique(y1)
oi=[]
for s in xi:
    oi.append(y.count(s))

Pi=[]
for i in range(len(xi)):
    prob=(math.comb(n,xi[i]))*(p**xi[i])*((1-p)**(n-xi[i]))
    Pi.append(prob)

Ei=[]
for i in range(len(xi)):
    Ei.append(1000*Pi[i])

oif=[]
Eif=[]
xif=[]
for i in range(len(xi)):
    if not(oi[i]>0 and Ei[i]<1e-6):       #check tolerance
        oif.append(oi[i])
        Eif.append(Ei[i])
        xif.append(xi[i])
               
test=[]
for i in range(len(xif)):
    test.append(((oif[i]-Eif[i])**2)/Eif[i])

chi_test=sum(test)
    
print("chi_test= ",chi_test)


#p-value
from scipy.stats import chi2
pvalue = 1-chi2.cdf(chi_test,len(xif)-1)
print("p_value= ", pvalue)


#test
if pvalue<0.05:
    print("Data doesn't follow Binomial")
else:
    print("Data follows binomial")
    

```

```python
#geometric(p)
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    count=1
    a=np.random.rand()
    while a>p:
        count+=1
        a=np.random.rand()
    y.append(count)

#test
y1=np.array(y)
xi=np.unique(y1)
oi=[]
for s in xi:
    oi.append(y.count(s))

Pi=[]
for i in range(len(xi)):
    prob=p*(1-p)**(xi[i]-1)
    Pi.append(prob)

Ei=[]
for i in range(len(xi)):
    Ei.append(1000*Pi[i])

oif=[]
Eif=[]
xif=[]
for i in range(len(xi)):
    if not(oi[i]>0 and Ei[i]<1e-6):      #check tolerance
        oif.append(oi[i])
        Eif.append(Ei[i])
        xif.append(xi[i])
               
test=[]
for i in range(len(xif)):
    test.append(((oif[i]-Eif[i])**2)/Eif[i])

chi_test=sum(test)
    
print("chi_test= ",chi_test)


#p-value
from scipy.stats import chi2
pvalue = 1-chi2.cdf(chi_test,len(xif)-1)
print("p_value= ", pvalue)


#test
if pvalue<0.05:
    print("Data doesn't follow Geometric")
else:
    print("Data follows Geometric")
    

```

```python
#negative binomial(r,p)
r=int(input("enter r: "))
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    x=[]
    for j in range(r):
        count=1
        a=np.random.rand()
        while a>p:
            count+=1
            a=np.random.rand()
        x.append(count)
    y.append(sum(x))

#test
y1=np.array(y)
xi=np.unique(y1)
oi=[]
for s in xi:
    oi.append(y.count(s))

Pi=[]
for i in range(len(xi)):
    prob=math.comb(xi[i]-1,r-1)*(p**r)*((1-p)**(xi[i]-r))
    Pi.append(prob)

Ei=[]
for i in range(len(xi)):
    Ei.append(1000*Pi[i])

oif=[]
Eif=[]
xif=[]

for i in range(len(xi)):
    if not(oi[i]>0 and Ei[i]<1e-6):      #check tolerance
        oif.append(oi[i])
        Eif.append(Ei[i])
        xif.append(xi[i])
               
test=[]
for i in range(len(xif)):
    test.append(((oif[i]-Eif[i])**2)/Eif[i])

chi_test=sum(test)
    
print("chi_test= ",chi_test)


#p-value
from scipy.stats import chi2
pvalue = 1-chi2.cdf(chi_test,len(xif)-1)
print("p_value= ", pvalue)


#test
if pvalue<0.05:
    print("Data doesn't follow negative binomial")
else:
    print("Data follows negative binomial")
    
```

```python
#poisson(lambda)

import numpy as np
import matplotlib.pyplot as plt
lambdaa=float(input("enter lambda: "))
y=[]
for i in range(1000):
    x=[]
    count=0
    a=np.random.exponential(1/lambdaa)
    x.append(a)
    while sum(x)<=1:
        count+=1
        a=np.random.exponential(1/lambdaa)
        x.append(a)
    y.append(count)


#test
y1=np.array(y)
xi=np.unique(y1)
oi=[]
for s in xi:
    oi.append(y.count(s))

Pi=[]
for i in range(len(xi)):
    prob = math.exp(-lambdaa) * (lambdaa**xi[i]) / math.factorial(xi[i])
    Pi.append(prob)

Ei=[]
for i in range(len(xi)):
    Ei.append(1000*Pi[i])

oif=[]
Eif=[]
xif=[]

for i in range(len(xi)):
    if not(oi[i]>0 and Ei[i]<1e-6):      #check tolerance
        oif.append(oi[i])
        Eif.append(Ei[i])
        xif.append(xi[i])
               
test=[]
for i in range(len(xif)):
    test.append(((oif[i]-Eif[i])**2)/Eif[i])
        
chi_test=sum(test)
    
print("chi_test= ",chi_test)


#p-value
from scipy.stats import chi2
pvalue = 1-chi2.cdf(chi_test,len(xif)-1)
print("p_value= ", pvalue)


#test
if pvalue<0.05:
    print("Data doesn't follow negative binomial")
else:
    print("Data follows negative binomial")
    
```
