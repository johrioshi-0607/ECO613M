```python
import numpy as np
```

```python
#A

p=0.5
mu1=1
sigsq1=1
mu2=2
sigsq2=3

y=np.array(np.random.binomial(1,p,10000))
x=[]
for i in y:
    if i==1:
        x.append(np.random.normal(mu1,np.sqrt(sigsq1)))
    else:
        x.append(np.random.normal(mu2,np.sqrt(sigsq2)))
        
mu1_new=0
mu2_new=0

def normal_pdf(x,mu,var):
    return (1/np.sqrt(2*np.pi*var))*np.exp(-(x-mu)**2/(2*var))

while(np.abs(mu1_new-mu1)>1e-6 or np.abs(mu2_new-mu2)>1e-6):     # or/and

    mu1_new=0
    mu2_new=0
    d1sum=0    
    for j in range(10000):
        f1 = normal_pdf(x[j],mu1,sigsq1)
        f2 = normal_pdf(x[j],mu2,sigsq2)
        
        d1=p*f1/(p*f1+(1-p)*f2)
        mu1_new+=d1*x[j]

        d1sum+=d1
        
        mu2_new+=(1-d1)*x[j]


    mu1_new/=d1sum
    mu2_new/=(10000-d1sum)

    sigsq1_new=0
    sigsq2_new=0
    for j in range(10000):

        f1 = normal_pdf(x[j],mu1,sigsq1)
        f2 = normal_pdf(x[j],mu2,sigsq2)

        d1 = p*f1/(p*f1+(1-p)*f2)

        sigsq1_new += d1*(x[j]-mu1_new)**2
        sigsq2_new += (1-d1)*(x[j]-mu2_new)**2

    sigsq1_new /= d1sum
    sigsq2_new /= (10000-d1sum)

    mu1=mu1_new
    mu2=mu2_new
    sigsq1=sigsq1_new
    sigsq2=sigsq2_new
    p = d1sum / 10000   #check


print("mu1= ",mu1,"mu2= ",mu2,"sigsq1= ",sigsq1,"sigsq2= ",sigsq2)

#check: estimates should be close to the initialized estimates
    
```
