```python
import numpy as np
p1=0.5
p2=0.3
p=0.5

y=np.random.binomial(1,p,10000)      ##
x=[]
for i in y:
    if i==1:
        x.append(np.random.binomial(20,p1))
    else:
        x.append(np.random.binomial(20,p2))

#----------------------------------------------------------------
import math

def binomial_pdf(x,p):
    return math.comb(20,x)*(p**x)*(1-p)**(20-x)    ##

p1_new=0
p2_new=0                                     ##

while np.abs(p1-p1_new)>1e-6 or np.abs(p2_new-p2)>1e-6:             ## or
    p1_new=0
    p2_new=0  
    gamma=np.zeros(10000)

    for j in range(10000):
        f1=binomial_pdf(x[j],p1)
        f2=binomial_pdf(x[j],p2)

        gamma[j] = (p*f1) / (p*f1 + (1-p)*f2)
        
    # M-step
    p = np.mean(gamma)

    
    x=np.array(x)


    p1_new = np.sum(gamma * x) / (20 * np.sum(gamma))
    p2_new = np.sum((1-gamma) * x) / (20 * np.sum(1-gamma))

    p1=p1_new
    p2=p2_new

print("p =", p)
print("p1 =", p1)
print("p2 =", p2)
```
