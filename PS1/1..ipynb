```python
#binomial(n,p)
import numpy as np
import matplotlib.pyplot as plt
n=int(input("enter n: "))
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    x=[]
    for j in range(n):
        a=np.random.rand()
        if a<p:
            x.append(1)
        else:
            x.append(0)
    y.append(x.count(1))

print(y[0:11])

plt.hist(y,bins=5, density=True)
plt.show()
```

```python
#geometric(p)
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    count=1
    a=np.random.rand()
    while a>p:
        count+=1
        a=np.random.rand()
    y.append(count)

print(y[0:11])

plt.hist(y,bins=5, density=True)
plt.show()
```

```python
#negative binomial(r,p)
r=int(input("enter r: "))
p=eval(input("enter p: "))
y=[]
for i in range(1000):
    x=[]
    for j in range(r):
        count=1
        a=np.random.rand()
        while a>p:
            count+=1
            a=np.random.rand()
        x.append(count)
    y.append(sum(x))

print(y[0:11])

plt.hist(y,bins=5, density=True)
plt.show()
```
